### Part 2: Apply Dimensionality Reduction

```python
#standardize dataset using a scalar
from sklearn.preprocessing import StandardScaler
myopia_scaled = StandardScaler().fit_transform(myopia)
print(myopia_scaled)
```

```python
# Using PCA to reduce dimensions of data

# Initialising PCA model - using 0.9 will preserve 90% of explained variance in dimensionality reduction
pca = PCA(n_components=0.9)

# Apply PCA
myopia_pca = pca.fit_transform(myopia_scaled)
```

```python
myopia_pca
```

```python
myopia_pca_df = pd.DataFrame(myopia_pca)
myopia_pca_df
```

```python
 # Initialize t-SNE model
tsne = TSNE(learning_rate='auto')
```

```python
tsne_features = tsne.fit_transform(myopia_pca) 
```

```python
tsne_features
```

```python
# Confirming that the dimensions have been reduced to 2:
tsne_features.shape 
```

```python
#Create a scatter plot of the output from t-SNE above

plt.scatter(tsne_features[:,0], tsne_features[:,1])
plt.show()
```

The plot above shows 4 or 5 distinct clusters, however we can't be sure just by looking. We need to use the elbow method to determine the number of clusters to use in our Kmeans model.

### Part 3: Perform a Cluster Analysis with K-means

```python
#initialise inertia with empty list & set range for K
inertias = []
k = list(range(1, 10))

# Using for loop to create a list of inertias for each K from 1 to 10 - we are also creating the kmeans model here
for number in k:
    km = KMeans(n_clusters=number, random_state=0)
    #we are using the tsne_features data
    km.fit(tsne_features)
    inertias.append(km.inertia_)

print(inertia)
```

```python
# create dataframe using K values & calculated inertia above
elbow_data = {"k": k, "inertia": inertias}
elbow_dataframe = pd.DataFrame(elbow_data)
elbow_dataframe.head()
```

```python
# Plot an elbow curve to find the best number of clusters to use
plt.plot(elbow_dataframe['k'], elbow_dataframe['inertia'])

plt.xticks(range(1,10))

#labelling x and y axes 
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')

#giving the plot a title 
plt.title('Elbow curve plot for MYOPIA data')

#showing the plot
plt.show()
```
